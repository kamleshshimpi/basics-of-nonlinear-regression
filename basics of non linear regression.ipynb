```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

linear regression with degree 1

```python
x = np.arange(5, 15, 0.1)
##You can adjust the slope and intercept to verify the changes in the graph
y = 2*(x) + 3
y_noise = 2*np.random.normal(size=x.size)
#y_noise = 1*np.random.normal(size=x.size)
data = y + y_noise

plt.plot(x, data,'bo')
plt.plot(x,y,'red' ) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()

```

non linear regression with degree 2 

```python
x = np.arange(5, 15, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 1*(x**2)+2*(x) + 3
y_noise = 15* np.random.normal(size=x.size)
data = y + y_noise

plt.plot(x, data,'bo')
plt.plot(x,y,'red' ) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

non linear regression with degree 3

```python
x=np.arange(-5,5,0.1)
y=(x**3)-(x**2)+2*(x)+2
noise=40*np.random.normal(size=x.size)
data=y+noise
plt.plot(x,data,'bo')
plt.plot(x,y,'red')
plt.show()
```

non linear regression with degree 4

```python
x=np.arange(-5,5,0.1)
y=(x**4)+(x**3)-(x**2)+2*(x)+2
noise=40*np.random.normal(size=x.size)
data=y+noise
plt.plot(x,data,'bo')
plt.plot(x,y,'red')
plt.show()
```

log function

```python
x=np.arange(-5,5,0.1)
y=np.log(x)
#noise=2*np.random.normal(size=x.size)
#data=y+noise
#plt.plot(x,data,'bo')
plt.plot(x,y,'bo')
plt.plot(x,y,'red')
plt.show()
```

exponential function

```python
x=np.arange(-5,5,0.1)
y=np.exp(x)
plt.plot(x,y,'bo')
plt.plot(x,y,'red')
```

sigmoid function

```python
#equation of sigmoid function is a+b/1+(c**(x-d)) where x is dataset
x=np.arange(-5,5,0.1)
a=1
b=4
c=3
d=2
#y=a+b/(1+np.power(c,x-d))
y=a-b/(1+np.power(c,x-d))
plt.plot(x,y,'bo')
plt.plot(x,y,'red')
```
